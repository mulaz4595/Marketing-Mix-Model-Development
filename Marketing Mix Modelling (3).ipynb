```python
#  Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
#  Loading the Data
# Make sure to upload the CSV file to your notebook/folder first
df = pd.read_csv('Advertising And Sales.csv')  # Change filename if different

# Display basic info about the dataset
print("Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
print("\nDataset info:")
df.head()
```

```python
# Explorind data
print("\n" + "="*50)
print("DATA EXPLORATION")
print("="*50)
```

```python
# Basic statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Checking column names
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
# Cell 4: Data Visualization
plt.figure(figsize=(15, 10))
```

```python
# Subplot 1: Sales distribution
plt.subplot(2, 3, 1)
plt.hist(df['Sales'], bins=20
         , color='blue', alpha=0.7)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
```

```python
# Subplot 2: TV vs Sales
plt.subplot(2, 3, 2)
plt.scatter(df['TV'], df['Sales'], alpha=0.6, color='red')
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Spend')
plt.ylabel('Sales')
```

```python
# Subplot 3: Radio vs Sales
plt.subplot(2, 3, 3)
plt.scatter(df['Radio'], df['Sales'], alpha=0.6, color='green')
plt.title('Radio Advertising vs Sales')
plt.xlabel('Radio Spend')
plt.ylabel('Sales')
```

```python
# Subplot 4: Newspaper vs Sales
plt.subplot(2, 3, 4)
plt.scatter(df['Newspaper'], df['Sales'], alpha=0.6, color='orange')
plt.title('Newspaper Advertising vs Sales')
plt.xlabel('Newspaper Spend')
plt.ylabel('Sales')
```

```python
# Subplot 5: Correlation heatmap
plt.subplot(2, 3, 5)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
```

```python
# Subplot 6: Total spend vs Sales
plt.subplot(2, 3, 6)
df['Total_Spend'] = df['TV'] + df['Radio'] + df['Newspaper']
plt.scatter(df['Total_Spend'], df['Sales'], alpha=0.6, color='purple')
plt.title('Total Spend vs Sales')
plt.xlabel('Total Marketing Spend')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()

print("Visualizations created successfully!")
```

```python
#  Building Marketing Mix Model
print("\n" + "="*40)
print("BUILDING MARKETING MIX MODEL")
print("="*40)
```

```python
# Preparing the data
X = df[['TV', 'Radio', 'Newspaper']]  # Marketing channels
y = df['Sales']  # Target variable
```

```python
# Creating and training the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Making predictions
y_pred = model.predict(X)
```

```python
# Calculating model performance
r2 = r2_score(y, y_pred)
print(f"Model R-squared: {r2:.4f}")
print(f"Model explains {r2*100:.2f}% of sales variance")
```

```python
# Getting model coefficients
coefficients = pd.DataFrame({
    'Channel': ['TV', 'Radio', 'Newspaper'],
    'Coefficient': model.coef_,
    'Intercept': [model.intercept_] * 3
})

print("\nModel Coefficients:")
print(coefficients)

print("\nModel created successfully!")
```
